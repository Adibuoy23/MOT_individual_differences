```python
# clear all the variables
%reset -f
import numpy as np
import sys
import turicreate as tc
import turicreate.aggregate as agg
import os
from os import walk
import fnmatch
import scipy.io
import pandas as pd
import chart_studio.plotly as py
import plotly.graph_objs as go
from scipy.stats import norm
import scipy.stats as st
import random
from plotly import tools
# Path variable (where the data is located)
myPath = '../data/processed_files/MOT_json_files/'
destPath='../data/model_output_files/'
#destPath='./MOT_correspondence_files_2/'
# Append the path variable to existing search path
sys.path.append(myPath)
# Get the file information in the directory
file_list = []
for root, dirs, files  in walk(destPath):
    for filename in files:
        if fnmatch.fnmatch(filename.lower(),"*correspondence*"):
            file_list.append(filename)
file_list = [s for s in file_list if "downsampled" not in s]
```

```python
file_list
```

```python
def compute_avg_cost_per_run(x):
    #return np.mean(np.sum(x['cost'], axis = 1))
    return np.mean(x['cost'])

def compute_avg_cost_per_frame_per_run(x):
    out = np.sum(x['cost'], axis = 1)
    return out 
    #return np.mean(x['cost'])
```

```python
MOT_corr_data = tc.SFrame.read_json(destPath+file_list[0], orient = 'records')
#MOT_corr_data['correspondences_final'] = MOT_corr_data.apply(lambda x: list(x['correspondences'][-1]))
#MOT_corr_data['predicted_accuracy_final'] = MOT_corr_data.apply(lambda x: x['predicted_accuracy'][-1])
MOT_corr_data['avg_cost'] = MOT_corr_data.apply(lambda x: compute_avg_cost_per_run(x))
MOT_corr_data['avg_cost_per_frame'] = MOT_corr_data.apply(lambda x: compute_avg_cost_per_frame_per_run(x))
MOT_corr_data = MOT_corr_data.groupby(['trial_Num_Original'], operations = {'avg_cost_per_trial':agg.MEAN('avg_cost'),
                                                                            'avg_cost_per_frame_per_trial':agg.MEAN('avg_cost_per_frame'),
                                                               'std_cost_per_trial':agg.STD('avg_cost'),
                                                                           'predicted_accuracy':agg.MEAN('predicted_accuracy'),
                                                                           'speed':agg.MEAN('speed'),
                                                                           'num_targets':agg.MEAN('num_targets')})
MOT_corr_data['subject_Num']=int(file_list[0][-8:-5])
for corr_file in file_list[1:]:
    temp = tc.SFrame.read_json(destPath+corr_file,orient='records')
    temp['avg_cost'] = temp.apply(lambda x: compute_avg_cost_per_run(x))
    temp['avg_cost_per_frame'] = temp.apply(lambda x: compute_avg_cost_per_frame_per_run(x))
    #temp['correspondences_final'] = temp.apply(lambda x: list(x['correspondences'][-1]))
    #temp['predicted_accuracy_final'] = temp.apply(lambda x: x['predicted_accuracy'][-1])
    temp = temp.groupby(['trial_Num_Original'],operations = {'avg_cost_per_trial':agg.MEAN('avg_cost'),
                                                             'avg_cost_per_frame_per_trial':agg.MEAN('avg_cost_per_frame'),
                                                             'std_cost_per_trial':agg.STD('avg_cost'),
                                                             'predicted_accuracy':agg.MEAN('predicted_accuracy'),
                                                                          'speed':agg.MEAN('speed'),
                                                                          'num_targets':agg.MEAN('num_targets')})
    temp['subject_Num']=int(corr_file[-8:-5])
    MOT_corr_data = MOT_corr_data.append(temp)

MOT_corr_data = MOT_corr_data.sort(['subject_Num','trial_Num_Original'])
```

```python
behavior_data = tc.SFrame.read_json(myPath+'behaviour_data',orient='records')
behavior_data = behavior_data[behavior_data['subject_Num']!=152]
behavior_data = behavior_data[behavior_data['subject_Num']!=153]
target_load_info = behavior_data['trial_Num_Original','num_targets']
#target_load_info = behavior_data.groupby('trial_Num_Original',operations = {'num_targets':agg.MEAN('num_targets')})
target_load_info = target_load_info.sort('trial_Num_Original')
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
sub_level_acc_data = behavior_data.groupby(['subject_Num'],operations={'avg_accuracy':agg.MEAN('accuracy')})
sub_level_acc_data = sub_level_acc_data.sort(['subject_Num'])
model_sub_level_acc_data = MOT_corr_data.groupby(['subject_Num'], operations = {'avg_predicted_accuracy':agg.MEAN('predicted_accuracy'), 'avg_exp_cost':agg.MEAN('avg_cost_per_trial')})
model_sub_level_acc_data = model_sub_level_acc_data.sort(['subject_Num'])
model_vs_sub_corr_comparison = sub_level_acc_data
model_vs_sub_corr_comparison['avg_predicted_accuracy'] = model_sub_level_acc_data['avg_predicted_accuracy']
model_vs_sub_corr_comparison['avg_accuracy_ranked'] = 51-scipy.stats.rankdata(sub_level_acc_data['avg_accuracy'].to_numpy(),method='dense')
model_vs_sub_corr_comparison['avg_predicted_ranked'] = 51-scipy.stats.rankdata(sub_level_acc_data['avg_predicted_accuracy'].to_numpy(),method='dense')
model_vs_sub_corr_comparison['avg_exp_cost'] = model_sub_level_acc_data['avg_exp_cost']
```

```python
from statsmodels.formula.api import ols
pd_sub_level_comparison = model_vs_sub_corr_comparison.to_dataframe()

mod = ols("avg_accuracy ~ avg_exp_cost", data=pd_sub_level_comparison)
res = mod.fit()
print(res.summary())
```

```python
from plotly.graph_objs import *
x = model_vs_sub_corr_comparison['avg_exp_cost'].to_numpy()
y = model_vs_sub_corr_comparison['avg_accuracy'].to_numpy()
sub_id = model_vs_sub_corr_comparison['avg_accuracy_ranked'].to_numpy().astype(str)
trace=[]
trace.append(go.Scatter(x = x, y = y,
                  mode='markers+text',
                  text = sub_id,
                  marker = dict(size = 25, color='orange',line=dict(width = 3.5, color='black'), opacity = 0.5),
                       showlegend=False))

x_lin_space = np.linspace(x.min(), x.max(),100)
y_lin_space = res.params[0] + res.params[1]*x_lin_space
trace.append(go.Scatter(x = x_lin_space, y = y_lin_space,
                       mode='lines',line=Line(color='gray', dash = 'dash', width = 7.5), showlegend=False,
                       ))

data = trace

annotations=[
        dict(
            x=40,
            y=.75,
            xref='x1',
            yref='y1',
            text='rsq :'+str(round(res.rsquared,3)),
            showarrow=False,
            ax=0,
            ay=-40
        ),
        dict(
        x = 33,
        y = .65,
        xref='x1',
        yref = 'y1',
        text = '$n = 50$',
        showarrow=False,
        ax = 0,
        ay = -40,
        )
    ]

fig = tools.make_subplots(rows=1, cols=1,)
fig['layout'].update(height = 800, width = 800, title='Good trackers make eye optimal movements <br> so that provide resistance against target drops ',
                     font = dict(size=20))

fig['layout']['annotations']=annotations

fig.append_trace(trace[0],1,1)
fig.append_trace(trace[1],1,1)

fig['layout']['xaxis1'].update(title='avg cost of eye movements (responsible for dropping targets)')
fig['layout']['yaxis1'].update(title='Participant accuracy ranked')

iplot(fig, filename='model-vs-human-accuracy-cost-correlation')
```

```python

trace = []
trace_fit = []



for i,tload in enumerate(sub_avg_cost_per_trial['num_targets'].unique().sort()):
    tload_data = sub_avg_cost_per_trial[sub_avg_cost_per_trial['num_targets']==tload]
    for j,speed in enumerate(tload_data['speed'].unique().sort()):
        condition_data = tload_data[tload_data['speed']==speed]
        condition_data = condition_data.groupby(['num_targets_chosen'],operations = {'avg_cost_per_trial':agg.MEAN('avg_cost_per_trial')})
        condition_data = condition_data.sort(['num_targets_chosen','avg_cost_per_trial'])
        trace.append(go.Scatter(x = condition_data['num_targets_chosen'].to_numpy(),
                               y = condition_data['avg_cost_per_trial'].to_numpy(),
                               mode = 'markers',
                               marker = dict(size = 12),
                               showlegend=False,
                               name = 'tload'+str(tload)+' speed:'+str(speed)))
        mod = ols("avg_cost_per_trial~num_targets_chosen", data=condition_data.to_dataframe())
        res = mod.fit()
        print(res.rsquared)
        trace_fit.append(go.Scatter(x = np.linspace(condition_data['num_targets_chosen'].min(),condition_data['num_targets_chosen'].max(),100),
                       y = np.linspace(condition_data['num_targets_chosen'].min(),condition_data['num_targets_chosen'].max(),100)*res.params[1] + res.params[0],
                        mode='lines',
                      line = dict(width = 7, color='gray', dash = 'dash'),
                        name = 'tload'+str(tload)+' speed:'+str(speed)+' fit',
                       showlegend=False))
trace_2d = list(np.array(trace).reshape(6,4))
trace_fit_2d = list(np.array(trace_fit).reshape(6,4))
```

```python
fig = tools.make_subplots(rows=6, cols=4)

for i in range(6):
    for j in range(4):
        fig.add_trace(trace_2d[i][j], i+1, j+1)
        fig.add_trace(trace_fit_2d[i][j], i+1, j+1)

fig['layout'].update(height=2000, width=1000, title='cost associated with the number of targets tracked')

fig['layout']['yaxis1'].update(title='Target load 3')
fig['layout']['yaxis5'].update(title='Target load 4')
fig['layout']['yaxis9'].update(title='Target load 5')
fig['layout']['yaxis13'].update(title='Target load 6')
fig['layout']['yaxis17'].update(title='Target load 7')
fig['layout']['yaxis21'].update(title='Target load 8')

fig['layout']['xaxis21'].update(title='Speed (2.8 deg/sec)')
fig['layout']['xaxis22'].update(title='Speed (5.6 deg/sec)')
fig['layout']['xaxis23'].update(title='Speed (8.4 deg/sec)')
fig['layout']['xaxis24'].update(title='Speed (11.2 deg/sec)')

py.iplot(fig, filename='cost associated with number of targets tracked')
```
